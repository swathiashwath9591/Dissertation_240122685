```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('Ocean_Acidification_Dataset.csv')

plt.figure(figsize=(8, 5))
plt.plot(df['Year'], df['CO2_Emissions_GT'], label='CO2 Emissions (GT)')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (Gigatons)')
plt.title('CO2 Emissions vs Year')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Year'], df['Atmospheric_CO2_ppm'], label='Atmospheric CO2 (ppm)', color='green')
plt.xlabel('Year')
plt.ylabel('Atmospheric CO2 (ppm)')
plt.title('Atmospheric CO2 vs Year')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Year'], df['Ocean_Surface_pH'], label='Ocean Surface pH', color='purple')
plt.xlabel('Year')
plt.ylabel('Ocean Surface pH')
plt.title('Ocean Surface pH vs Year')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Year'], df['Sea_Surface_Temperature_C'], label='Sea Surface Temp (°C)', color='red')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.title('Sea Surface Temperature vs Year')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
